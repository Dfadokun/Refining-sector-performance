```python
#import required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
#plt.style.use('seaborn')


%matplotlib inline
```

```python
pd.set_option('max_columns', None)
pd.set_option('max_rows',500)
pd.set_option('display.width', None)
```

Import all the required data

```python
#import refining capacity data
ref_capacity = pd.read_excel('Refining capacity.xlsx',index_col="Unnamed: 0")
ref_capacity.rename(columns={"Unnamed: 0":''},inplace=True)
#transpose data
ref_capacity=ref_capacity.T

ref_capacity=ref_capacity.replace(0,np.NaN)
ref_capacity
```

```python
global_ref_thruput = pd.read_excel('global Refining throughput.xlsx',index_col="Unnamed: 0")
#Transpose data
global_ref_thruput=global_ref_thruput.T
global_ref_thruput
```

```python
#import crude production capacity data
prod_capacity = pd.read_excel('Production capcity.xlsx',index_col="Unnamed: 0")
#prod_capacity.rename(columns={"Unnamed: 0":'Country'},inplace=True)
#Transpose Data
prod_capacity=prod_capacity.T
prod_capacity
```

```python
#Petroleum production consumption for OPEC countries
consumption = pd.read_excel('Pet prod consumption.xlsx')
consumption.rename(columns={'Unnamed: 0':''},inplace=True)
consumption.set_index('',inplace=True)
#Transpose Data
consumption=consumption.T
consumption
```

```python
#Extract Nigeria consumption data
consumption_nig=consumption.iloc[:,54:60]
consumption_nig.columns=['Total', 'Gasoline', 'Kerosene', 'Distillates', 'Residuals','Others']
consumption_nig=consumption_nig.replace('na',np.NaN)
consumption_nig=consumption_nig.iloc[20:,:]
consumption_nig
```

```python
#OPEC country refinng thruput
refining_thruput = pd.read_excel('Refining throughput.xlsx')
refining_thruput
```

```python
#Extract Nigeria data from OPEC country refinng thruput 
nigeria_thruput = refining_thruput[refining_thruput['Country']=='Nigeria']
nigeria_thruput.set_index(['Product'],inplace=True)
nigeria_thruput.drop("Country",axis=1,inplace=True)

#Transpose Data
nigeria_thruput=nigeria_thruput.T
nigeria_thruput
```

```python
#Import global petroleum production data
consumption1 = pd.read_excel('Pet prod consumption1.xlsx')
consumption1.rename(columns={'Unnamed: 0':''},inplace=True)
consumption1.set_index('',inplace=True)
consumption1=consumption1.T
consumption1
```
